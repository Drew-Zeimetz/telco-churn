## Key findings from data understanding

- TotalCharges is stored as a string — should be numeric, needs investigating

- Only 3 numeric columns: SeniorCitizen, tenure, MonthlyCharges

## Investigating TotalCharges
TotalCharges should be numeric like MonthlyCharges but is stored as a string.
Converting it and checking for hidden nulls.

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
print(f"TotalCharges nulls: {df['TotalCharges'].isnull().sum()}")
df[df['TotalCharges'].isnull()]
```

## Finding: 11 missing TotalCharges values

All 11 rows with missing TotalCharges have tenure = 0, therefore, these are brand 
new customers who have not been billed yet. This is not bad data, it is 
a business logic gap. Since they have no billing history, we will fill 
these with 0.

```python
df['TotalCharges'] = df['TotalCharges'].fillna(0)

print(f"TotalCharges nulls after fix: {df['TotalCharges'].isnull().sum()}")

print(f"TotalCharges dtype: {df['TotalCharges'].dtype}")
```
